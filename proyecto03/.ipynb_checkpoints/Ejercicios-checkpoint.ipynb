```python
#__IMPORTAMOS TODAS LAS LIBRERIAS NECESARIAS__
import pandas as pd
import csv
import requests
import re
import ast
import matplotlib
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
%matplotlib inline
```

```python
APIurl = "http://numbersapi.com/1..100"
APIresp = requests.get(APIurl)
print(APIresp.ok)
print(APIresp.status_code)
print(APIresp.elapsed)
APIresp.text
```

```python
# Parsar string a un array (matrix, lista de listas)
x = ast.literal_eval(APIresp.text)
print(x)
type(x)
print(x.keys())
```

```python
# Convertimos los datosen un dataframe.
df= pd.DataFrame([x.keys(), x.values()]).T
df.columns= ['keys', 'values'] 
df
```

```python
# Grabamos el resultado como CSV 
CSVfilename = "ApiEjercicio.csv"
df.to_csv(CSVfilename, encoding='utf-8', index=False, quoting=csv.QUOTE_ALL)
```

```python
# Creamos una columna nueva con la longitud de la columna values.
df['longitud'] = df['values'].apply(len)
```

```python
df
```

```python
# Hacemos un histograma de la columna longitud.
plt.hist(df["longitud"], density=True, bins=30)
plt.ylabel('Probabilidad')
plt.xlabel('longitud')
plt.show()
```

```python
def Ran():
    
    r = RandomW()
    df =pd.DataFrame(r)
    for i in range(0, 100):
        df[i] = RandomW()
    
    return df

```
