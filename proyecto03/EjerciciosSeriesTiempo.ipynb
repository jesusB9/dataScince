# Ejercicio Serie de tiempo


- Crear una serie de tiempo con los siguientes 3 componentes:
- Tendencia Y0 = m * X + b, siendo M = 0.3 y b = 10
- Estacionalidad Y1 = sin(0.1 * X)
- Volatilidad Y2 = random, siendo random un número aleatorio (distribución, media y desviación a elegir)
- Dibuja en pantalla la serie de tiempo de los 100 primeros valores de X (entre 0 y 100)

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
# Le pasamos los paramatres
M = 0.3
b = 10
X = np.arange(0, 100)
```

```python
# Tendencia Lineal
Y0 = M * X  + b
plt.plot(Y0)
plt.show()
```

```python
# Estacionalidad
Y1 = np.sin(0.1 * X)
plt.plot(Y1)
plt.show()
```

```python
# Volatilidad
Y2 = np.random.normal(0, 10, 100)
plt.plot(Y2)
plt.show()
```

```python
Y = Y0 + Y1 + Y2
plt.plot(X, Y)
plt.show()
```
